```python
# Retreiving all the dubai hotels urls 
hotel_url=[]
start=0
end=39
for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//a[@class="a78ca197d0"]')
    for i in url:
        hotel_url.append(i.get_attribute('href'))
    next_button=driver.find_element(By.XPATH,'//div[@class="b16a89683f cab1524053"]')
    next_button.click()
    time.sleep(3)
url=driver.find_elements(By.XPATH,'//a[@class="a78ca197d0"]')
for i in url:
    hotel_url.append(i.get_attribute('href'))
```

```python
name=[]
city=[]
location=[]
num_reviews=[]
user_ratings=[]
star_rating=[]
review_score=[]
free_wifi=[]
family_rooms=[]
non_smoking_rooms=[]
restaurant=[]
bar=[]
heating=[]
lift=[]
b_c1=[]
b_c2=[]
b_c3=[]
room_type=[]
single_bed=[]
double_bed=[]
prices=[]
```

```python
for i in hotel_url:
    driver.get(i)
    time.sleep(3)
    
    #For hotel name
    try:
        Name=driver.find_element(By.XPATH,'//div[@data-capla-component-boundary="b-property-web-property-page/PropertyHeaderName"]/h2')
        name.append(Name.text)
    except NoSuchElementException :
        name.append(np.nan)
    except StaleElementReferenceException :
        name.append(np.nan)
     
    #For city 
    #As we already know the city
    try:
        city.append('Dubai')
    except NoSuchElementException :
        city.append(np.nan)
    except StaleElementReferenceException :
        city.append(np.nan)
        
     
    #For location
    try:
        Location=driver.find_element(By.XPATH,'//p[@class="address address_clean"]/span')
        location.append(Location.text)
    except NoSuchElementException :
        location.append(np.nan)
    except StaleElementReferenceException :
        location.append(np.nan)
        
       #For number of reviews
    try:
        Num_reviews=driver.find_element(By.XPATH,'//div[@class="daaa8ff09f f62fdb6838"]/div/div[2]/span[2]')
        num_reviews.append(Num_reviews.text.split(' ')[2])
    except NoSuchElementException :
        num_reviews.append(np.nan)
    except StaleElementReferenceException :
        num_reviews.append(np.nan)
    except IndexError :
        num_reviews.append(np.nan)
    
    #For user ratings
    try:
        User_ratings=driver.find_element(By.XPATH,'//div[@class="daaa8ff09f f62fdb6838"]/div/div')
        user_ratings.append(User_ratings.text)
    except NoSuchElementException :
        user_ratings.append(np.nan)
    except StaleElementReferenceException :   
        user_ratings.append(np.nan)
    #For star ratings
    try:
        Star_rating=driver.find_element(By.XPATH,'//div[@class="afa1dec5e7"]/div/div/div/div/div/div/div[2]')
        star_rating.append(Star_rating.text.split('out')[0].split('as')[1])
    except NoSuchElementException :
        star_rating.append(np.nan)
    except StaleElementReferenceException :
        star_rating.append(np.nan)
    except IndexError :
        star_rating.append(np.nan)
       
    #For review score
    try:
        Review_score=driver.find_element(By.XPATH,'//div[@data-testid="host-profile"]/header/div[2]/div/span')
        review_score.append(Review_score.text.split(':')[1].split(' ')[1])
    except NoSuchElementException :
        review_score.append(np.nan)
    except StaleElementReferenceException : 
        review_score.append(np.nan)
    except IndexError :
        review_score.append(np.nan)
    #For free wifi
    try:
        Free_wifi=driver.find_element(By.XPATH,'//div[@class="fa47acd48c"]/div/div')
        if Free_wifi.text.__contains__('Free WiFi'):
            free_wifi.append('Yes')
        else :
            free_wifi.append('No')
    except NoSuchElementException :
        free_wifi.append(np.nan)
    except StaleElementReferenceException :
        free_wifi.append(np.nan)
    #For family_rooms
    try:
        Family_rooms=driver.find_element(By.XPATH,'//div[@class="fa47acd48c"]/div/div')
        if Family_rooms.text.__contains__('Family rooms'):
            family_rooms.append('Yes')
        else :
            family_rooms.append('No')
    except NoSuchElementException :
        family_rooms.append(np.nan)
    except StaleElementReferenceException :  
        family_rooms.append(np.nan)
        #For non_smoking_rooms
    try:
        Non_smoking_rooms=driver.find_element(By.XPATH,'//div[@class="fa47acd48c"]/div/div')
        if Non_smoking_rooms.text.__contains__('Non-smoking rooms'):
            non_smoking_rooms.append('Yes')
        else :
            non_smoking_rooms.append('No')
    except NoSuchElementException :
        non_smoking_rooms.append(np.nan)
    except StaleElementReferenceException :
        non_smoking_rooms.append(np.nan)

    #For restaurant
    try:
        Restaurant=driver.find_element(By.XPATH,'//div[@class="e50d7535fa"]')
        if Restaurant.text.__contains__('Restaurant'):
            restaurant.append('Yes')
        else :
            restaurant.append('No')
    except NoSuchElementException :
        restaurant.append(np.nan)
    except StaleElementReferenceException :
         restaurant.append(np.nan)
    #For Bar
    try:
        Bar=driver.find_element(By.XPATH,'//div[@class="e50d7535fa"]')
        if Bar.text.__contains__('Bar'):
            bar.append('Yes')
        else :
            bar.append('No')
    except NoSuchElementException :
        bar.append(np.nan)
    except StaleElementReferenceException :
        bar.append(np.nan)
    #For Heating
    try:
        Heating=driver.find_element(By.XPATH,'//div[@class="e50d7535fa"]')
        if Heating.text.__contains__('Heating'):
            heating.append('Yes')
        else :
            heating.append('No')
    except NoSuchElementException :
        heating.append(np.nan)
    except StaleElementReferenceException :
         heating.append(np.nan)
    #For lift
    try:
        Lift=driver.find_element(By.XPATH,'//div[@class="e50d7535fa"]')
        if Lift.text.__contains__('Lift'):
            lift.append('Yes')
        else :
            lift.append('No')
    except NoSuchElementException :
        lift.append(np.nan)
    except StaleElementReferenceException :
        lift.append(np.nan)
    
    #For Breakfast Cuisine1
    try:
        B_c1=driver.find_element(By.XPATH,'//div[@class="property-highlights ph-icon-fill-color"]/div[2]/div[2]')
        b_c1.append(B_c1.text)
       
    except NoSuchElementException :
        b_c1.append(np.nan)
    except StaleElementReferenceException :
        b_c1.append(np.nan)
    
    #For Breakfast Cuisine2
    try:
        B_c2=driver.find_element(By.XPATH,'//div[@class="property-highlights ph-icon-fill-color"]/div[2]/div[2]')
        b_c2.append(B_c2.text)
       
    except NoSuchElementException :
        b_c2.append(np.nan)
    except StaleElementReferenceException :
        b_c2.append(np.nan)
    
    #For Breakfast Cuisine3
    try:
        B_c3=driver.find_element(By.XPATH,'//div[@class="property-highlights ph-icon-fill-color"]/div[2]/div[2]')
        b_c3.append(B_c3.text)
       
    except NoSuchElementException :
        b_c3.append(np.nan)
    except StaleElementReferenceException :
         b_c3.append(np.nan)
    
    #For Room type
    try:
        Room_type=driver.find_element(By.XPATH,'//div[@class="hprt-block"]/div/a/span')
        room_type.append(Room_type.text)
       
    except NoSuchElementException :
        room_type.append(np.nan)
    except StaleElementReferenceException :
        room_type.append(np.nan)
    
    #For single_bed
    try:
        Single_bed=driver.find_element(By.XPATH,'//td[@class="hprt-table-cell -first hprt-table-cell-roomtype droom_seperator"]')
        if Single_bed.text.__contains__('single bed'):
            single_bed.append('Yes')
        else :
            single_bed.append('No')
    except NoSuchElementException :
        single_bed.append(np.nan)
    except StaleElementReferenceException :
        single_bed.append(np.nan)
    
     #For double_bed
    try:
        Double_bed=driver.find_element(By.XPATH,'//td[@class="hprt-table-cell -first hprt-table-cell-roomtype droom_seperator"]')
        if Double_bed.text.__contains__('double bed'):
            double_bed.append('Yes')
        else :
            double_bed.append('No')
    except NoSuchElementException :
        double_bed.append(np.nan)
    except StaleElementReferenceException :
         double_bed.append(np.nan)
    
     #For prices
    try:
        Prices=driver.find_element(By.XPATH,'//span[@class="prco-valign-middle-helper"]')
        prices.append(Prices.text)
       
    except NoSuchElementException :
        prices.append(np.nan)
    except StaleElementReferenceException :
        prices.append(np.nan)
```
